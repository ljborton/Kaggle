```python
def prime1(max_num):
    prime=[2]

    for i in range(3,max_num):
        for j in prime:
            if (i%j==0 & i!=j):
                break 
            if j>i/2:
                prime.append(i)
                break
    print(prime)
```

```python
def prime2(max_num):
    prime=[2]

    for i in range(3,max_num):
        for j in prime:
            if (i%j==0 & i!=j):
                break 
            if j>i**0.5:
                prime.append(i)
                break
    return(prime)
```
